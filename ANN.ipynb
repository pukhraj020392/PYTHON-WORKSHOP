```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset  = pd.read_csv("Churn_Modelling.csv")
dataset
```

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
le = LabelEncoder()
dataset['Geography'] = le.fit_transform(dataset['Geography'])
dataset['Gender'] = le.fit_transform(dataset['Gender'])
```

```python
dataset.head()
```

```python
dataset.dtypes
```

```python
x = dataset.iloc[:,3:13].values
x
```

```python
y = dataset.iloc[:,13].values
y
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.25, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
x_train.shape
```

```python
import keras

from keras.models import Sequential
from keras.layers import Dense
```

```python
clf = Sequential()

clf.add(Dense(output_dim=6, init='uniform', activation='relu', input_dim=10)) #1st hidden layer

clf.add(Dense(output_dim=6, init='uniform', activation='relu')) #2nd hidden layer

clf.add(Dense(output_dim=3, init='uniform', activation='relu')) #3nd hidden layer
              
clf.add(Dense(output_dim=1, init='uniform', activation='sigmoid')) # output layer
                          
```

```python
clf.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
clf.fit(x_train, y_train, batch_size=10, nb_epoch=50)
```

```python
y_pred = clf.predict(x_test)
y_pred
```

```python
y_pred=(y_pred>0.5)

from sklearn.metrics import confusion_matrix
from sklearn import metrics

cm = confusion_matrix(y_test, y_pred)
```
